# Provincial Dataset

```python
wait_prov.columns
```

```python
wait_prov.loc[wait_prov['50th percentile/Days'].isna()]
```

```python
wait_prov = wait_prov.dropna(subset=['50th percentile/Days'])
```

```python
wait_prov.sort_values('Indicator', inplace = True)
```

```python
wait_prov.head()
```

```python
#wait_prov['50th percentile/Days'] = wait_prov['50th percentile/Days'].fillna(0)
#wait_prov['50th percentile/Days'] = wait_prov['50th percentile/Days'].astype(np.int64)
#wait_prov['50th percentile/Days'].replace(0, np.nan, inplace=True)
```

```python
indicator_list = wait_prov['Indicator'].unique()
indicator_list
```

```python
#rescale mean data so we can take average across all procedures

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

wait_prov['wait_time_scaled'] = np.nan

for ind in indicator_list:
    to_be_scaled = wait_prov.loc[wait_prov['Indicator']== ind, '50th percentile/Days']
    to_be_scaled = np.array(to_be_scaled)
    to_be_scaled = np.reshape(to_be_scaled,(-1, 1))
    wait_time_scaled = scaler.fit_transform(to_be_scaled)
    wait_prov.loc[wait_prov['Indicator'] == ind, 'wait_time_scaled'] = wait_time_scaled
```

```python
#Test codes
to_be_scaled = wait_prov.loc[wait_prov['Indicator']== 'Radiation Therapy', '50th percentile/Days']
to_be_scaled = np.array(to_be_scaled)
to_be_scaled = np.reshape(to_be_scaled,(-1, 1))
wait_time_scaled = scaler.fit_transform(to_be_scaled)
wait_time_scaled            
```

```python
#compare the data in df with test codes to make sure the scale data are correctly appended
wait_prov.loc[wait_prov['Indicator'] == 'Radiation Therapy', 'wait_time_scaled']
```

```python
prov_list = wait_prov['Province/territory'].unique()
prov_list
```

```python
#exclude_list = ["Hip Replacement","Knee Replacement"]

include_list = np.delete(indicator_list, np.argwhere((indicator_list == "Hip Replacement") & (indicator_list == "Knee Replacement")))
wait_prov['avg_wait_time_exclude'] = np.nan
wait_prov['avg_wait_time_all'] = np.nan
column_names = ['Province/territory', "avg_wait_time_exclude", "avg_wait_time_all"]
prov_WT = pd.DataFrame(columns = column_names)

for prov in prov_list:
    to_be_averaged = wait_prov.loc[(wait_prov['Indicator'].isin(include_list) ) &
                                   (wait_prov['Province/territory']== prov),
                                   'wait_time_scaled']
    to_be_averaged = np.array(to_be_averaged)
    avg1 = np.nanmean(to_be_averaged)
    
    to_be_averaged = wait_prov.loc[wait_prov['Province/territory']== prov,
                                   'wait_time_scaled']
    to_be_averaged = np.array(to_be_averaged)
    avg2 = np.nanmean(to_be_averaged)
    
    wait_prov.loc[wait_prov['Province/territory'] == prov, 'avg_wait_time_exclude'] = avg1
    wait_prov.loc[wait_prov['Province/territory'] == prov, 'avg_wait_time_all'] = avg2
    prov_WT.loc[prov_WT.shape[0]] = [prov, avg1, avg2]
```

```python
prov_WT
```

```python
to_be_averaged = wait_prov.loc[(wait_prov['Indicator'].isin(include_list)) & 
                               (wait_prov['Province/territory']== "Saskatchewan") , 'wait_time_scaled']
to_be_averaged = np.array(to_be_averaged)
display(to_be_averaged)
average = np.nanmean(to_be_averaged)
print(average)
```

```python
wait_region['wait_time_scaled'] = np.nan

for ind in indicator_list:
    to_be_scaled = wait_region.loc[wait_region['Indicator']== ind, '50th percentile/Days']
    to_be_scaled = np.array(to_be_scaled)
    to_be_scaled = np.reshape(to_be_scaled,(-1, 1))
    wait_time_scaled = scaler.fit_transform(to_be_scaled)
    wait_region.loc[wait_region['Indicator'] == ind, 'wait_time_scaled'] = wait_time_scaled
```

```python
wait_region = wait_region.dropna(subset=['HR_UID'])
```

```python
hr_uid = wait_region['HR_UID'].unique()
#hr_uid_str = ["%.2f" % x for x in hr_uid]
include_list = ["Hip Replacement", "Knee Replacement"]
wait_region['avg_wait_time_replacement'] = np.nan

column_names = ['HR_UID','Province/territory', 'Region', "avg_wait_time_replacement"]
reg_WT = pd.DataFrame(columns = column_names)


for uid in hr_uid:
    province = wait_region.loc[wait_region['HR_UID'] == uid,'Province/territory'].values
    region = wait_region.loc[wait_region['HR_UID'] == uid,'Region'].values
    to_be_averaged = wait_region.loc[(wait_region['Indicator'].isin(include_list) ) &
                                     (wait_region['HR_UID'] == uid),
                                     'wait_time_scaled']
    to_be_averaged = np.array(to_be_averaged)
    avg = np.nanmean(to_be_averaged)
    wait_region.loc[wait_region['HR_UID'] == uid, 'avg_wait_time_replacement'] = avg
    reg_WT.loc[reg_WT.shape[0]] = [uid, province[0], region[0], avg]
```
